***Compresión de imagenes por factorización SVD***




En este proyecto utilizaremos un algoritmo que comprima una imagen conservando la calidad de ella mediante el método de facturación SVD desarrollamos un algoritmo utilizando el lenguaje de programación Python logrando así una solución a nuestra pregunta planteada.

La descomposición en valores singulares (SVD) es una forma de factorizar cualquier matriz en tres matrices diferentes. Es una generalización de la descomposición de valores propios que  solo funciona en matrices cuadradas.El SVD describe una matriz X con tres componentes de matriz usando esta ecuación: $X=U*S*V^T$


Imagen1.png

Imagen 1: El SVD descompone cualquier matriz en tres nuevas matrices, U, S y $V^T$. La matriz S se denomina matriz Singular, y las matrices U y $V^T$ se denominan matrices singulares izquierda y derecha, respectivamente.

image.png

Imagen 2: Los valores singulares son las raíces cuadradas de los valores propios de la matriz cuadrada $X*X^T o  X^T*X$. La matriz singular contiene los valores singulares de la matriz X a lo largo de la diagonal y ceros en otros lugares. Los valores singulares están dispuestos en orden descendente a lo largo de la diagonal desde la esquina superior izquierda a la esquina inferior derecha.

image.png

Imagen 3:Las columnas de U se llaman vectores singulares izquierdos. Estos son los vectores propios de la matriz $X*X^T$. Están ordenados de izquierda a derecha en orden descendente de valor singular.

image.png

Imagen 4: Las filas de $V^T$  se denominan vectores singulares rectos. Estos son los vectores propios de la matriz $X^T*X$

image.png

Imagen 5: Las filas de $V^T$  se denominan vectores singulares rectos. Estos son los vectores propios de la matriz $X^T*X$.La matriz original X se puede calcular usando la combinación lineal de columnas de U, valores singulares en S y filas de $V^T$.

Una de las aplicaciones más comunes utilizando el método de factorización SVD son:


*   En los motores de búsqueda como: Google.
*   Base de reconocimiento facial: por ejemplo, Facebook.
*   Sistemas de recomendación como Amazon y Netflix.
*   Una herramienta estándar para la exploración de datos y la reducción de dimensiones en genómica.
*   En la Compresión de imagenes.  


**Objetivo general:**Complementar un algoritmo en Python el cual por medio de la factorización SVD permita minimizar el espacio requerido para almacenar una imagen conservando la calidad de ella.

***¿Es posible hacer que una imagen reduzca su peso de manera drástica sin que pierda calidad?***

Se puede utilizar la SVD en la compresión de imágenes digitales para disminuir la cantidad de información a transmitir.
Supongamos que una matriz X de $m*n$, representa una imagen en tonos de grises de tamaño $m*n$ pixeles. Cada entrada de A contiene un numero que representa el valor del tono de gris de la imagen entre 0 y 255, también puede ser entre 0.0 y 1.0. Interpretando 0(0.0) como negro y 255(1.0) como blanco y valores intermedios como tonos de gris. 

Dada la SVD de X

$X = U_1*S_1*(V_1)^T + … + U_r*S_r*(V_r)^T$

Podemos encontrar $K < r$ tal que

$X_K = U_1*S_1*(V_1)^T + … + U_K*S_K*(V_K)^T$

Siendo esta la aproximación de X aquella que representa una imagen nítida de menor peso. A continuación, se ve lo anterior de manera grafica:
